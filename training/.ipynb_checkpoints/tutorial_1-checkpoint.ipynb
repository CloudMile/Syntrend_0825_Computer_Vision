```python
import tensorflow as tf, numpy as np, pandas as pd
```

### 設定檔

```python
n_epoch = 200
learning_rate = 0.001
```

### 準備資料 
+ outerX: 亂數產生
+ outerY: 根據function y = 0.5x + 0.3 產生的資料, 與 outerX有線性關係

```python
outerX = np.random.random(10000)
outerY = 0.5 * outerX + 0.3
outerX, outerY
```

```python
tf.reset_default_graph()
```

### Placeholder
+ 接收外部參數

```python
# 接收 outerX data
placeholderX = tf.placeholder(tf.float32, shape=[None])
# 接收 outerY data
placeholderY = tf.placeholder(tf.float32, shape=[None])
```

### Variable
+ tensorflow內部專用變數, 與 Placeholder 最大分別就是資料不是外部傳入

```python
varW = tf.Variable(tf.random_uniform(shape=[1]), tf.float32)
varB = tf.Variable(tf.random_uniform(shape=[1]), tf.float32)
varY = varW * placeholderX + varB
```

### Loss Function 與 Optimizer
+ 利用MSE(均方誤差) 去逼進傳進來的 placeholderY

```python
# 均方誤差
loss = tf.losses.mean_squared_error(placeholderY, varY)
# 使用 AdamOptimizer
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
# 最小化 loss, 此時會 backpropagation 去調整 varW, varB的值
trainer = optimizer.minimize(loss)
```

### 使用tf.Session與TensorFlow溝通, 觀察Variables變化 => varW, varB
+ feed_dict 傳入資料
+ session run trainer 會 backpropagation 調整 varW, varB 的值

```python
# with tf.Session() as sess:
sess = tf.Session()
# 初始化所有Variable
init = tf.global_variables_initializer()
sess.run(init)
for i in range(n_epoch):
    _, w_, b_ = sess.run([trainer, varW, varB], feed_dict={placeholderX: outerX, placeholderY: outerY})
    print('varW: {}, varB: {}'.format(w_, b_) )
    
sess.close()
```

### 整合在一起

```python
n_epoch = 200
learning_rate = 0.0001

tf.reset_default_graph()
with tf.Graph().as_default() as g:
    # 接收 outerX data
    placeholderX = tf.placeholder(tf.float32, shape=[None])
    # 接收 outerY data
    placeholderY = tf.placeholder(tf.float32, shape=[None])
    
    varW = tf.Variable(tf.random_uniform(shape=[1]), tf.float32)
    varB = tf.Variable(tf.random_uniform(shape=[1]), tf.float32)
    varY = varW * placeholderX + varB
    
    # 均方誤差
    loss = tf.losses.mean_squared_error(placeholderY, varY)
    # 使用 AdamOptimizer
    optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
    # 最小化 loss, 此時會 backpropagation 去調整 varW, varB的值
    trainer = optimizer.minimize(loss)
    with tf.Session() as sess:
        init = tf.global_variables_initializer()
        sess.run(init)
        for i in range(n_epoch):
            _, w_, b_ = sess.run([trainer, varW, varB], feed_dict={placeholderX: outerX, placeholderY: outerY})
            print('varW: {}, varB: {}'.format(w_, b_) )
```
